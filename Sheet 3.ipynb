```python
import pandas as pd
import numpy as np
df_movie=pd.read_excel(r'C:\Users\solan\OneDrive\Desktop\Sample\Movie data.xlsx')
```

```python
df_p1=df_movie[['Title','Budget','Gross Revenue']]
```

```python
pd.options.display.float_format = '${:,.2f}'.format
df_p1.pivot_table(df_p1, index='Title',margins =True, margins_name = 'Grande Total').head(20)
```

```python
df_p2=df_movie[['Genre','Rating','Gross Revenue']]
```

```python
df_p2.pivot_table(df_p2,index='Genre',columns='Rating',aggfunc={'Gross Revenue':np.sum}).replace(np.nan,'$0')
```

```python
df_p3=df_movie[['Genre','Rating','Gross Revenue']]
```

```python
df_p3.pivot_table(df_p3, index=['Genre','Rating'],aggfunc=np.sum,margins=True,margins_name="Total")
```

```python
df_p4=df_movie[['Genre','Rating','Gross Revenue']]
```
